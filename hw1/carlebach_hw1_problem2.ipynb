```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# See myTorch.py for my basic implementation of neural net
# Cite:  loosely inspired by Joel Grus video, https://www.youtube.com/watch?v=o64FV-ez6Gw
from myTorch import SequentialModel
```

```python
# Read data, clean in data, extract features and response variable
df = pd.read_csv("Housing_Data.csv")
badrows = df.index[df['floorArea'] == "sqft"].tolist()
df.drop(badrows, inplace=True)
floor_area = pd.to_numeric(df.floorArea).to_numpy()
bedrooms = df.bedrooms.to_numpy()
price = df.value.to_numpy()
```

```python
# Look at data before we model
# Plot price vs. floor area
fig, ax = plt.subplots(1, 1, figsize=(10,8))
ax.set_title("Price vs. floor area", fontsize=18)
ax.set_xlabel("floor area", fontsize=14)
ax.set_ylabel("price", fontsize=14)
ax.scatter(floor_area, price)
ax.grid(True)
plt.show()
```

```python
# Drop observations with floor area > 7500
idx = np.where(floor_area > 7500)
floor_area = np.delete(floor_area, idx)
bedrooms = np.delete(bedrooms, idx)
price = np.delete(price, idx)

# Look again
fig, ax = plt.subplots(1, 1, figsize=(10,8))
ax.set_title("Price as function of floor_area and bedrooms", fontsize=18)
ax.set_xlabel("floor area", fontsize=14)
ax.set_ylabel("price", fontsize=14)
for bedroom in range(1,8):
    idx = np.where(bedrooms == bedroom)
    ax.scatter(floor_area[idx], price[idx], label=f"bedrooms={bedroom}")

ax.grid(True)
ax.legend()
plt.show()
```

```python
# Put data into X & y numpy arrays and scale
scaler = StandardScaler()
X = np.zeros([price.shape[0], 2])
X[:,0] = floor_area
X[:,1] = bedrooms
scaler.fit(X)
X = scaler.transform(X)
y = price / 1000

# Create train and test data
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=.2, random_state=89)
```

```python
# Create NN model using myTorch implementation
model = SequentialModel(n_inputs=2, hidden_dim=2)
print(model.summary())

# Fit the model
history = model.fit(X_train=X_train, y_train=y_train,
                    X_val=X_val, y_val=y_val,
                    n_epochs=100, lr=1e-6, 
                    batch_size=128, verbose=False)
```

```python
# Plot results
fig, ax = plt.subplots(1, 1, figsize=(10,8))
ax.set_title("Training and Validation Loss per Epoch", fontsize=18)
ax.set_xlabel("epoch", fontsize=14)
ax.set_ylabel("rmse (000)", fontsize=14)
ax.plot(np.array(history["loss"]) ** .5, label = "train loss")
ax.plot(np.array(history["val_loss"]) ** .5, label = "val loss")
ax.legend(fontsize=14)
ax.grid(True)
plt.show()
```

```python
price_mean = price.mean()
print(f"Mean price = ${price_mean :6.0f}")

rmse = min(history["val_loss"]) ** 0.5
print(f"RMSE = ${rmse :3.0f}000")
```

**Commnet:**

- The RMSE on validation data is pretty high relative to the mean price of a home.
- This does not seem like a very good model.
